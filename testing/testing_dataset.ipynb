```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import sys
sys.path.insert(0, '../')
from operations import adjust_dataset, split_data, Ann
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../docs/data/GlobalWeatherRepository.csv')
target_variable = 'temperature_celsius'

df = adjust_dataset(df, target_variable=target_variable)
```

```python
# lets see the correlation between variables (not including the target variable)
# and plot the correlation matrix

df = df.drop(columns=['location_name'])

# lets see the correlation between variables

corr_matrix = df.corr()
print(corr_matrix[target_variable].sort_values(ascending=False))

# lets plot the correlation matrix
fig, ax = plt.subplots(figsize=(10, 10))
ax.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.show()






```
